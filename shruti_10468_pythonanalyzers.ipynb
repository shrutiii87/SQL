```python
# Import all required libraries

import numpy as np
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Load dataset correctly

df = pd.read_csv("Telecom_Customer_Support_Tickets_Gigantic_Dataset - Telecom_Customer_Support_Tickets_Gigantic_Dataset.csv.csv")
print("dataset load succesfully")
```

```python
# TASK :- 2 Data cleaning & EDA
```

```python
# Identify and handle missing values

print("Missing values in each coloumns")

print("\nTHIS DETECT MISSING VALUES:-\n")
print(df.isnull().sum())

print("\nTHIS DETECT NON MISSING VALUES:-\n")
print(df.notnull().sum())

print("\nREMOVE MISSING VALUES:-\n")
print(df.dropna())

print("\nFILLING MISSING VALUES:-\n")
print(df.fillna(0))
```

```python
df['Resolution_Time_Hours'] = df['Resolution_Time_Hours'].fillna(df['Resolution_Time_Hours'].median())
df['Satisfaction_Score'] = df['Satisfaction_Score'].fillna(df['Satisfaction_Score'].median())
print("Missing Values After Handling:\n")
print(df.isnull().sum())
```

```python
# Detect and handle duplicate records

print(df.nunique())
print("\ndrop duplicates :- to remove duplicates\n")
print(df.drop_duplicates)
```

```python
# Generate

# Summary Statistics
df.describe().T
```

```python
# HISTOGRAM (RESOLUTION TIME DISTRIBUTION)

plt.figure(figsize=(7, 5))
sns.histplot(df['Resolution_Time_Hours'], bins=30, kde=True, color='skyblue')
plt.title('Resolution Time Distribution')
plt.xlabel('Resolution Time (Hours)')  
plt.ylabel('Frequency')              
plt.show()
```

```python
# SCATTER PLOT

plt.figure()
plt.scatter(df['Resolution_Time_Hours'],
df['Satisfaction_Score'])
plt.xlabel('Resoultion hours')
plt.ylabel('Satisfaction score')
plt.title("resoultion time vs customer satisfaction")

plt.show()
```

```python
# PIE PLOT

issue_counts=df['Issue_Category'].value_counts()
plt.figure()
plt.pie(issue_counts,labels=issue_counts.index)
plt.title("distribution of issue categories")
plt.show()
```

```python
df["Created_Date"] = pd.to_datetime(df["Created_Date"])
yearly_data = df.groupby(df["Created_Date"].dt.year)["Resolution_Time_Hours"].mean()

plt.figure()
plt.plot(yearly_data.index, yearly_data.values)
plt.xlabel("Year")
plt.ylabel("Average Resolution Time (Hours)")
plt.title("Average Resolution Time Over Years")

plt.show()
```
